## Exploratory Analysis

When beginning data analysis, we oftern wish to explore data to get to know it better. This includes understanding something about the features of our data (e.g. distribution, range, median, mean, categories) and about the relationship these variables may have with time, or other variables in our dataset. From exploratory analysis, we might identify outliers, needed data transformations, or what predictive models may apply.

### BarPlots

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 1)
data = Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot(kind='bar', ax=axes[0])
data.plot(kind='barh', ax=axes[1])
```

### Histogram/Density Plots

```python
comp = np.random.normal(0, 1, size=500)
values = Series(comp)
values.hist(bins=100, alpha=0.3, color='k', normed=True)
values.plot(kind='kde', style = 'k--')
```
